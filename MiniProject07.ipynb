```python
'''
Example WorkFlow !!!
Prices: [100, 101, 102, 103, 102, 104, 106, 108, 107, 105]
Short SMA (3): [...]
Long SMA (5): [...]
Buy signal at day 4 price 102
Sell signal at day 7 price 108
Total Profit: 6.0

'''
```

```python
print("Simple Moving Average Crossover Strategy")
prices = [100, 101, 102, 103, 102, 104, 106, 108, 107, 105,107,108,110,109,103]
short_window = 3
long_window = 5
short_sma = []
long_sma = []

for i in range(len(prices) - short_window + 1):
    window = prices[i:i+short_window]
    short_sma.append(sum(window) / short_window)
    
for i in range(len(prices) - long_window + 1):
    window = prices[i:i+long_window]
    long_sma.append((sum(window)) / long_window)
    

for i in  range(len(long_sma)):
    if(short_sma[i] > long_sma[i]) & (short_sma[i-1] < long_sma[i]):
        print(f"Buy Signal on Day {i}")
    elif(short_sma[i+2] < long_sma[i]) & (long_sma[i-1] < short_sma[i]):
        print(f"Sell Signal on Day {i}")
        
print("")    
print(f"Short Term Sma Values {short_sma}")
print("")
print(f"Long Term Sma Values {long_sma}")
        
```

```python
import matplotlib.pyplot as plt

plt.plot(short_sma, label="Short Sma", marker="o")
plt.plot(long_sma, label="Long Sma", marker="s")
plt.xlabel("Date")
plt.ylabel("Values")
plt.title("Simple Moving Averages")
plt.legend()


plt.show()
```
